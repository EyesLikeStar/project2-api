## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
text = requests.get('https://en.wikipedia.org/wiki/Greater_China').text
text_m = bs4.BeautifulSoup(text,'html.parser').get_text()
print(text_m)
```

```python
#download an image with beautifulsoup and save it in this repository
soup = bs4.BeautifulSoup(text, 'html.parser')
image_tag = soup.find('img')
image_url = image_tag['src']
if not image_url.startswith('http'):
    image_url = 'https://en.wikipedia.org' + image_url
image_response = requests.get(image_url)
with open('python_logo.svg', 'wb') as file:
    file.write(image_response.content)
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
links = soup.find_all('a', href=True)
for link in links[:10]:
    print(link['href'][:100])
```

```python

text_content = []
for link in scripts['link']:
    try:
        response = requests.get(link)
        soup = bs4.BeautifulSoup(response.content, 'html.parser')
        text = soup.get_text()[:1000]  # Extract first 1000 characters of text
        text_content.append(text)
    except Exception as e:
        text_content.append('Error downloading content')

# Add a new column to the DataFrame with the downloaded text
scripts['script_text'] = text_content
```

## Part 4: TMDB database

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 

apikey = '35e0d99d0da6b453be328723a4e9e931'
url = 'https://api.themoviedb.org/3/movie/now_playing'
params = {
    'api_key': apikey,
    'language': 'en-US',
    'page': 2
}
response = requests.get(url, params=params)
movies = response.json().get('results', [])

movie_data = []
for x in movies:
    movie_info = {
        'title': x.get('title'),
        'backdrop_path': x.get('backdrop_path'),
        'popularity': x.get('popularity'),
        'poster_path': x.get('poster_path'),
    }
    movie_data.append(movie_info)
df = pd.DataFrame(movie_data)
df.to_csv('movies_in_theaters.csv', index=False)
```
